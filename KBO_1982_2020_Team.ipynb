```python
# 필요한 라이브러리 import
from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd
import time
```

# 팀 기록실 - 타자 데이터

```python
# setup Driver | Chrome: 크롬 드라이버를 사용하는 driver 생성
driver = webdriver.Chrome('./chromedriver')  # Chrome driver 경로
driver.implicitly_wait(10)

result_bat = []
for i in range(1982, 2021):
    count = 0
    driver.get('http://www.statiz.co.kr/stat.php?opt=0&sopt=0&re=0&ys=' + str(i) + '&ye=' + str(i) + '&se=0&te=&tm=&ty=0&qu=auto&po=0&as=&ae=&hi=&un=&pl=&da=1&o1=WAR_ALL_ADJ&o2=TPA&de=1&lr=5&tr=&cv=&ml=1&sn=30&si=&cn=')
    time.sleep(1)
    
    html = driver.page_source
    soup = BeautifulSoup(html, 'html.parser')
    tr = soup.select('table#mytable > tbody > tr')
    
    for r in tr:
        row = []
        td = r.find_all('td')
        for d in td:
            if d == td[2]:
                year = d.text.strip()
                if int(year) > 81: year = '19' + year
                else: year = '20' + year
                row.append(year)
                
            else:
                row.append(d.text.strip())
        
        if len(row) > 0:
            count += 1
            result_bat.append(row)
    
    print(i, '년: ', count, sep='')
                    
data_bat = pd.DataFrame(result_bat)
```

```python
data_bat.head()
```

```python
dataset_bat = data_bat.drop(data_bat.columns[3], axis='columns')
dataset_bat.head()
```

```python
dataset_bat.columns = ['No', 'Name', 'Year', 'G', 'PA', 'AB', 'R', 'H', '2B', '3B', 'HR', 'TB', 'RBI', 'SB', 'CS', 'BB', 'HBP', 'IBB', 'SO', 'GDP', 'SAC', 'SF', 'AVG', 'OBP', 'SLG', 'OPS', 'wOBA', 'wRC+', 'WAR', 'WPA']
dataset_bat.head()
```

# 팀 기록실 - 투수 데이터

```python
# setup Driver | Chrome: 크롬 드라이버를 사용하는 driver 생성
driver = webdriver.Chrome('./chromedriver')  # Chrome driver 경로
driver.implicitly_wait(10)

result_pit = []
for i in range(1982, 2021):
    count = 0
    driver.get('http://www.statiz.co.kr/stat.php?opt=0&sopt=0&re=1&ys=' + str(i) + '&ye=' + str(i) + '&se=0&te=&tm=&ty=0&qu=auto&po=0&as=&ae=&hi=&un=&pl=&da=1&o1=WAR&o2=OutCount&de=1&lr=5&tr=&cv=&ml=1&sn=30&si=&cn=')
    time.sleep(1)
    
    html = driver.page_source
    soup = BeautifulSoup(html, 'html.parser')
    tr = soup.select('table#mytable > tbody > tr')
    
    for r in tr:
        row = []
        td = r.find_all('td')
        for d in td:
            if d == td[2]:
                year = d.text.strip()
                if int(year) > 81: year = '19' + year
                else: year = '20' + year
                row.append(year)
                
            else:
                row.append(d.text.strip())
        
        if len(row) > 0:
            count += 1
            result_pit.append(row)
    
    print(i, '년: ', count, sep='')
                    
data_pit = pd.DataFrame(result_pit)
```

```python
dataset_pit = data_pit.drop(data_pit.columns[3], axis='columns')
dataset_pit.head()
```

```python
dataset_pit.columns = ['No', 'Name', 'Year', 'G', 'CG', 'ShO', 'GS', 'W', 'L', 'SV', 'HLD(HD)', 'IP', 'R', 'ER', 'TBF', 'H', '2B', '3B', 'HR', 'BB', 'IBB', 'HBP', 'SO', 'BK', 'WP', 'ERA', 'FIP', 'WHIP', 'ERA+', 'FIP+', 'WAR', 'WPA']
dataset_pit.head()
```

# 타자 데이터 + 투수 데이터

```python
dataset_bat['for_merge'] = dataset_bat['Name'] + dataset_bat['Year']
dataset_pit['for_merge'] = dataset_pit['Name'] + dataset_pit['Year']
dataset_pit = dataset_pit.drop(['No', 'Name', 'Year'], axis='columns')
dataset = pd.merge(dataset_bat, dataset_pit, how='inner', on='for_merge', suffixes=('_B', '_P'), copy=True)
```

# 팀 순위 데이터

```python
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import Select
```

```python
driver = webdriver.Chrome('./chromedriver')  # Chrome driver 경로
driver.implicitly_wait(10)
driver.get('https://www.koreabaseball.com/TeamRank/TeamRank.aspx')
time.sleep(2)

result_rank = []

for y in range(1982, 2021):
    count = 0
    selYear = Select(driver.find_element(By.ID, "cphContents_cphContents_cphContents_ddlYear"))
    selYear.select_by_value(str(y))
    time.sleep(1)
    
    html = driver.page_source
    soup = BeautifulSoup(html, 'html.parser')
    tr = soup.select('div#cphContents_cphContents_cphContents_udpRecord > table.tData > tbody > tr')

    for r in tr:
        row = []
        td = r.find_all('td')
        row.append(td[0].text.strip())  # 순위
        row.append(td[1].text.strip())  # 팀명
        row.append(y)
        row.append(td[2].text.strip())  # 경기 수
        row.append(td[3].text.strip())  # 승
        row.append(td[4].text.strip())  # 패
        row.append(td[5].text.strip())  # 무
        row.append(td[6].text.strip())  # 승률
        
        if len(row) > 0:
            count += 1
            result_rank.append(row)
    
    print(y, '년: ', count, sep='')
                    
data_rank = pd.DataFrame(result_rank)
```

```python
data_rank.head()
```

```python
data_rank.columns = ['Rank', 'Team', 'Year', 'G', 'W', 'L', 'D', 'P']
data_rank.head()
```

----

# 선수 Stats + 전적 + 승률 + 순위 데이터

```python
dataset_rank = data_rank.copy()
dataset_rank = dataset_rank.astype({'Year': str})
dataset_rank['for_merge'] = dataset_rank['Team'] + dataset_rank['Year']
dataset_rank = dataset_rank.drop(['Team', 'Year'], axis='columns')
dataset_rank.head()

dataset = dataset.merge(dataset_rank, how='inner', on='for_merge', suffixes=('_P', '_T'), copy=True)
dataset.head()
```

```python
dataset = dataset.drop(['for_merge'], axis='columns')
```

```python
dataset.to_csv('KBO_1982_2020_Team_Stats_Rank_EN.csv', encoding='cp949', index=False)
dataset.head()
```
